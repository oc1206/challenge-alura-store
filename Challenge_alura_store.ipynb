## **Extracción de datos**

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## **Análisis de datos**

### **Ingreso total por cada tienda**

```python
strListaDeTiendas = ["Tienda", "Tienda 2", "Tienda 3", "Tienda 4"]
listaDeTiendas = [tienda, tienda2, tienda3, tienda4]

ingresosTotalesPorTienda = [float(df['Precio'].sum()) for df in listaDeTiendas]

for i, ingreso in enumerate(ingresosTotalesPorTienda):
    print(f"El ingreso total de la tienda {i+1} es: {ingreso}")
```

```python
sIngresosTotalesPorTienda = pd.Series(ingresosTotalesPorTienda, index=strListaDeTiendas)
sIngresosTotalesPorTienda.plot(kind="bar", figsize=(6, 2), title="Ingresos totales por tienda")
```

#### Conclusión1: Podemos observar que la tienda4 es aquella que tiene menores ventas, mientras que la tienda 1 tiene las mayores ventas, teniendo una diferencia de 112,504,700 millones de la moneda utilizada.

### **Ventas por categoría**

```python
ventasPorCategoriaTienda = tienda['Categoría del Producto'].value_counts().sort_values(ascending=True).rename("Ventas")
ventasPorCategoriaTienda.plot(kind="line", figsize=(16, 4), title="Ventas por categoría Tienda 1")
ventasPorCategoriaTienda
```

```python
ventasPorCategoriaTienda2 = tienda2['Categoría del Producto'].value_counts().sort_values(ascending=True).rename("Ventas")
ventasPorCategoriaTienda2.plot(kind="pie", figsize=(8, 4), title="Ventas por categoría Tienda 2")
ventasPorCategoriaTienda2
```

```python
ventasPorCategoriaTienda3 = tienda3['Categoría del Producto'].value_counts().sort_values(ascending=True).rename("Ventas")
ventasPorCategoriaTienda3.plot(kind="barh", figsize=(8, 4), title="Ventas por categoría Tienda 3")
ventasPorCategoriaTienda3
```

```python
ventasPorCategoriaTienda4 = tienda4['Categoría del Producto'].value_counts().sort_values(ascending=True).rename("Ventas")
ventasPorCategoriaTienda4.plot(kind="line", figsize=(16, 4), title="Ventas por categoría Tienda 4")
ventasPorCategoriaTienda4
```

#### Conclusión2: Podemos observar que las categorias más vendidas son muebles y electrónica, mientras los menos vendidos son instrumentos musicales y artículos del hogar.

### **Valoración media por tienda**

```python
calificacionesMediasPorTienda = [round(float(df["Calificación"].mean()), 2) for df in listaDeTiendas]

for i, calificacion in enumerate(calificacionesMediasPorTienda):
    print(f"La calificación promedio de la tienda {i+1} es: {calificacion}")
```

```python
sCalificacionesMediasPorTienda = pd.Series(calificacionesMediasPorTienda, index=strListaDeTiendas)
sCalificacionesMediasPorTienda.plot(kind="bar", figsize=(6, 2), title="Calificación promedio por tienda")
```

Conclusión3: Podemos observar la mejor reputación en la tienda 4, podría decir que talvez se cumplen más peticiones de los clientes respecto a los productos, lo cual puede implicar devoluciones, por tal motivo tengan menor reporte de ventas. sin embargo la reputacion de las 4 tiendas no varia mucho, la tienda 4 en relación con la tienda 1 quien tiene mejores ventas, la diferencia es de apenas 0.02ptos.

### **Valor del envío promedio por tienda**

```python

costosEnvioPromedio = [round(float(df["Costo de envío"].mean()), 2) for df in listaDeTiendas]
costosEnvioPromedio
```
